### Question Number 1
Which is the greater indicator of heart attack? BMI or Heart_Rate?

### Psuedocode
1. Convert the target variable (`Heart_Attack_Occurrence`) into numerical format (0 = No, 1 = Yes)
2. Compute the correlation of BMI and Heart Rate with `Heart_Attack_Occurrence`
3. Calculate the mean and standard deviation of BMI and Heart Rate for people who had a heart attack and those who did not
4. Compare the correlation values and statistical differences to determine which feature is a stronger indicator
5. Return the feature with the stronger association

### Single Line

```python
# Converting the target variable into 0 or 1

data["Heart_Attack_Occurrence"] = data["Heart_Attack_Occurrence"].map({"No": 0, "Yes": 1})
```

```python
# Computing the Correlation of BMI and Heart_Rate with the target variable

correlation_bmi = data["BMI"].corr(data["Heart_Attack_Occurrence"])
correlation_hr = data["Heart_Rate"].corr(data["Heart_Attack_Occurrence"])
print(correlation_bmi, correlation_hr)
```

```python
# Compute mean and standard deviation for each category

bmi_stats = data.groupby("Heart_Attack_Occurrence")["BMI"].agg(["mean", "std"])
hr_stats = data.groupby("Heart_Attack_Occurrence")["Heart_Rate"].agg(["mean", "std"])
```

```python
# Print correlation and statistical results

print(f"Correlation with Heart Attack Occurrence: BMI: {correlation_bmi} Heart Rate: {correlation_hr}")
```

```python
# Return the stronger indicator
if abs(correlation_bmi) > abs(correlation_hr):
    print("BMI is a stronger indicator of heart attack.")
else:
    print("Heart Rate is a stronger indicator of heart attack.")
```

### Function / Example Usage

```python
def compare_heart_attack_indicators(data):
    data["Heart_Attack_Occurrence"] = data["Heart_Attack_Occurrence"].map({"No": 0, "Yes": 1})

    # Computing the correlations
    correlation_bmi = data["BMI"].corr(data["Heart_Attack_Occurrence"])
    correlation_hr = data["Heart_Rate"].corr(data["Heart_Attack_Occurrence"])

    # Computing the mean and standard deviation for each category
    bmi_stats = data.groupby("Heart_Attack_Occurrence")["BMI"].agg(["mean", "std"])
    hr_stats = data.groupby("Heart_Attack_Occurrence")["Heart_Rate"].agg(["mean", "std"])

    # Printing results
    print(f"Correlation with Heart Attack Occurrence: BMI: {correlation_bmi} Heart Rate: {correlation_hr}")

    # Returning the stronger indicator
    if abs(correlation_bmi) > abs(correlation_hr):
        print("BMI is a stronger indicator of heart attack.")
    else:
        print("Heart Rate is a stronger indicator of heart attack.")
```

#### Example Usage

```python
compare_heart_attack_indicators(data)
```

### Question Number 1: 
When we look at the top 5 players and the bottom 5 players (determined by the rank), how they do compare in terms of the difference between max exit velocity and average velocity? * In baseball, a small difference between max and average velocity is desired because it means a player is consistent.

### Pseudocode
1. Extract the top 5 players and the bottom 5 players.
2. Calculate the difference between max velocity and average velocity for each player. 
3. Calculate the average difference for two groups: top 5 and bottom 5
4. Compare the two numbers

### Single Line

```python
# Extracting the top 5 players and the bottom 5 players
top5 = second_data.head(5)
bottom5 = second_data.tail(5)
```

```python
# Calculaing the difference between max velocity and average velocity for each player

top5["velocity_diff"] = top5["max_ev"] - top5["average_ev"]
bottom5["velocity_diff"] = bottom5["max_ev"] - bottom5["average_ev"]
```

```python
# Calculating the average difference for the two groups
top5_avg_diff = top5["velocity_diff"].mean()
bottom5_avg_diff = bottom5["velocity_diff"].mean()
```

```python
# Returning the results for comparison (result shows that top 5 players are more consistent in terms of exit velocity)

print(f"Top 5 Players - Avg Velocity Difference: {top5_avg_diff}")
print(f"Bottom 5 Players - Avg Velocity Difference: {bottom5_avg_diff}")
```

### Function / Example Usage

```python
def average_velo_diff (data):
    top5 = data.head(5)
    bottom5 = data.tail(5)

    # Calculateing the velocity difference
    top5["velocity_diff"] = top5["max_ev"] - top5["average_ev"]
    bottom5["velocity_diff"] = bottom5["max_ev"] - bottom5["average_ev"]

    # Calculating the average difference
    top5_avg_diff = top5["velocity_diff"].mean()
    bottom5_avg_diff = bottom5["velocity_diff"].mean()

    # Printing results for comparison
    if top5_avg_diff > bottom5_avg_diff:
        print("Top 5 Players are more inconsistent in terms of exit velocity")
    else:
        print("Top 5 Players are more consistent in terms of exit velocity")
```

#### Example Usage

```python
average_velo_diff(second_data)
```

### Question Number 2:
Is there a relationship between the number of batted ball events (times a player made contact) and the number of hard-hit balls (95+ mph exit velocity)?

### Pseudocode
1. Group the data by player
2. Aggregate the hard-hit counts per player
3. Create a data frame combining batted ball events and hard-hit counts
4. Compute the correlation between batted ball events and hard-hit balls
5. Print the correlation value and interpret the result

### Single Line

```python
# Grouping by player and taking the maximum batted ball events per player
player_batted_ball = second_data.groupby("player")["batted_ball_events"].max()
print(player_batted_ball)
```

```python
# Aggregating the hard-hit counts per player
player_hard_hit_95 = second_data.groupby("player")["hard_hit_95mph+"].sum()
```

```python
# Creating a DataFrame combining batted ball events and hard-hit counts
player_stats = pd.DataFrame({
    "batted_ball_events": player_batted_ball, 
    "hard_hit_95+": player_hard_hit_95
})

# Filling missing values with 0 in case some players have no hard-hit balls
player_stats = player_stats.fillna(0)
```

```python
# Computing the correlation
correlation_bbe_hh95 = player_stats["batted_ball_events"].corr(player_stats["hard_hit_95+"])
print(f"Correlation between Batted Ball Events and Hard-Hit Balls (95+ mph): {correlation_bbe_hh95}")
```

```python
# Interpretation
if abs(correlation_bbe_hh95) > 0.5:
    print("There is a strong relationship between total batted ball events and hard-hit balls.")
elif abs(correlation_bbe_hh95) > 0.2:
    print("There is a weak relationship between total batted ball events and hard-hit balls.")
else:
    print("There is little to no relationship between total batted ball events and hard-hit balls.")
```
